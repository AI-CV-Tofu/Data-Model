```python
import boto3
import zipfile
import io

# S3 클라이언트 설정
s3 = boto3.client('s3')

# 필요한 변수들
bucket_name = "square-tofu"  # S3 버킷 이름
zip_file_key = "CV_tofu.v3i.yolov11.zip"  # S3에서 ZIP 파일의 경로
upload_prefix = "7class_640/"  # 압축 해제 후 S3에 저장될 경로

# 1. S3에서 ZIP 파일을 메모리로 읽기
response = s3.get_object(Bucket=bucket_name, Key=zip_file_key)
zip_file_content = response['Body'].read()
print(f"ZIP 파일 {zip_file_key}을(를) 메모리로 로드 완료")

# 2. ZIP 파일 압축 해제
with zipfile.ZipFile(io.BytesIO(zip_file_content), 'r') as zip_ref:
    for file_name in zip_ref.namelist():
        # 3. 각 파일을 S3에 업로드
        file_content = zip_ref.read(file_name)
        s3_file_key = f"{upload_prefix}{file_name}"
        
        s3.put_object(Bucket=bucket_name, Key=s3_file_key, Body=file_content)
        print(f"파일 {file_name}을(를) S3 {s3_file_key}에 업로드 완료")

print("모든 파일 압축 해제 및 업로드 완료!")
```
