```python
from ultralytics import YOLO
import torch

# YOLO 모델 로드
model = YOLO("../yolo11s.pt")

# 디바이스 설정
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# PyTorch에서 디바이스 출력
print("gpu :", device)
```

```python
model.train(
    data="/home/ec2-user/SageMaker/7class_640_re/data.yaml",  # 데이터 경로
    epochs=50,  # 학습 epoch 수
    imgsz=640,  # 이미지 크기
    batch=16,  # 배치 크기
    device=0,  # GPU (device=0은 첫 번째 GPU를 사용)
    project="yolov11s",  # 프로젝트 이름
    name="7class_640_re"  # 실험 이름 (결과를 구분하기 위함)
)
```
