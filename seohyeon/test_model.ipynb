```python
from ultralytics import YOLO

# 학습된 모델 로드
model = YOLO("yolov11s/2class_6403/weights/best.pt")
```

```python
# 테스트 데이터 예측
results = model.predict(source="./2class_640/valid/images", save=True)
```

```python
# 예측 결과 출력
print(results)

# 결과 이미지 확인
import matplotlib.pyplot as plt
plt.imshow(results[0].plot())  # 첫 번째 이미지의 결과 표시
plt.show()
```
