```python
from ultralytics import YOLO
import torch
import os
import matplotlib.pyplot as plt

# YOLO 모델 로드
model = YOLO("../yuna/yolov11s/7class_640_re15/weights/best.pt")
```

```python
input_folder = "test_img/"  # 입력 폴더 경로
```

```python
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):  # 이미지 파일 필터링
        image_path = os.path.join(input_folder, filename)
        
        # YOLO 추론 수행
        results = model.predict(image_path)
        
        # 시각화
        plt.imshow(results[0].plot())  # 첫 번째 이미지의 결과 표시
        plt.title(f"Detection: {filename}")  # 이미지 이름 표시
        plt.axis('off')  # 축 제거
        plt.show()
```
