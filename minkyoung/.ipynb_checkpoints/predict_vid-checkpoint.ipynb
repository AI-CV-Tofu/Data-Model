```python
from ultralytics import YOLO
import cv2
from IPython.display import display, Image, clear_output

# Load a pretrained YOLO model
model = YOLO("./yolov11s/7class_640/weights/best.pt")

# Define path to video file
source = "두부.mp4"

# Run inference on the source
results = model(source, stream=True)

# Loop through each frame in the results
for result in results:
    frame = result.orig_img  # Original frame from the video
    annotated_frame = result.plot()  # Annotated frame with detections
    
    # Convert the frame to displayable format
    _, buffer = cv2.imencode('.jpg', annotated_frame)
    
    # Set the display size (width=640 here)
    display_frame = Image(data=buffer.tobytes(), width=400)  
    
    # Display the frame in the notebook
    clear_output(wait=True)  # Clear the previous frame
    display(display_frame)
```
