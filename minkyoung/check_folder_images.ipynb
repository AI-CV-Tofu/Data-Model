```python
import os
from PIL import Image
from IPython.display import display, clear_output

#    
folder_path = '/home/ec2-user/SageMaker/7class_640_re/valid/images'

#       
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

#   
for image_file in image_files:
    #  
    image_path = os.path.join(folder_path, image_file)
    
    #  
    img = Image.open(image_path)
    
    #    
    width, height = img.size
    resized_img = img.resize((width // 2, height // 2))
    
    #     
    clear_output(wait=True)
    
    #   
    print(f"File Name: {image_file}")
    display(resized_img)
    
    #      
    input("Press Enter to see the next image...")
```
